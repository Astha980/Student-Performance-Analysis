```python
df.drop_duplicates(inplace=True)
```

```python
parent_scores = df.groupby("parental level of education")[["math score","reading score","writing score"]].mean()
parent_scores
```

```python
testprep = df.groupby("test preparation course")[["math score","reading score","writing score"]].mean()
testprep
```

```python
corr = df[["math score","reading score","writing score"]].corr()
corr
```

```python
gender_scores = df.groupby("gender")[["math score","reading score","writing score"]].mean()

gender_scores
```

```python
df["Total Score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
)
df["Total Score"].describe()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    x="parental level of education",
    y="math score",
    data=df
)
plt.xticks(rotation=45)
plt.title("Math Scores by Parental Education")
plt.xlabel("Parental Education")
plt.ylabel("Math Score")
plt.savefig("Boxplot_Parental_Education.png")
plt.show()
```

```python
testprep.plot(kind="bar", figsize=(8,5))

plt.title("Test Preparation Comparison")
plt.ylabel("Average Score")

plt.savefig("TestPrep_Bar.png")
plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.savefig("Correlation_Heatmap.png")

plt.show()
```

```python
gender_scores.plot(kind="bar", figsize=(8,5))

plt.title("Gender vs Subject Performance")
plt.ylabel("Average Score")

plt.savefig("Gender_Subject.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["Total Score"],
    bins=20
)

plt.title("Total Score Distribution")
plt.xlabel("Total Score")
plt.ylabel("Number of Students")

plt.savefig("TotalScore_Histogram.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["reading score"],
    df["math score"]
)

plt.title("Reading vs Math Scores")
plt.xlabel("Reading Score")
plt.ylabel("Math Score")

plt.savefig("Reading_vs_Math.png")

plt.show()
```
